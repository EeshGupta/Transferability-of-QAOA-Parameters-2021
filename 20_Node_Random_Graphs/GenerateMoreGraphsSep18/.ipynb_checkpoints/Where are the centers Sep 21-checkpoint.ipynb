## For classifying centers

```python
def even_percent(graph):
  """
  Determines the relative percent of even degree nodes
  """
  nodes = graph.number_of_nodes()
  tot_ev = 0
  for i in range(nodes):
    if graph.degree[i]%2 == 0:
      tot_ev+=1
  return tot_ev/nodes
```

```python
from scipy.stats import gaussian_kde
from matplotlib import cm
from matplotlib.colors import Normalize

def get_centers(graphs_idx, g_data = g_data):
  '''
  Displays optima of graph indices specified
  '''
  gammas = []
  betas = []
  for i in graphs_idx:
    data = g_data[i]
    for j in range(len(data)):
      gamma = data[j][0]%1
      beta = data[j][1]%0.5
      gammas.append(gamma)
      betas.append(beta)
  ####################################################
  #Graphing 
  fig , ax = plt.subplots()
  # Generate fake data
  x = betas
  y = gammas

  # Calculate the point density
  xy = np.vstack([x,y])
  z = gaussian_kde(xy)(xy)


  # Sort the points by density, so that the densest points are plotted last
  idx = np.argsort(z)
  new_x = [x[i] for i in idx]
  new_y = [y[i] for i in idx]
  new_z = [z[i] for i in idx]
  #x, y, z = x[idx], y[idx], z[idx]

  #fig, ax = plt.subplots()
  ax.scatter(new_x, new_y, c=new_z, s=100)
  #plt.colorbar()
  # cbar = plt.colorbar(label = "Number of graph pairs per marker")



  norm = Normalize(vmin = np.min(z), vmax = np.max(z))
  cbar = fig.colorbar(cm.ScalarMappable(norm = norm), ax=ax)
  cbar.ax.set_ylabel('Number of graph optima per marker')

  ######################

  #####################
  ax.set_ylabel(r"$\gamma/2\pi$")#, fontsize = 20)
  ax.set_xlabel(r"$\beta/2\pi$")#, fontsize = 20)

  ####
  centers = [0 for i in range(6)]
  total = 0
  for i in range(len(gammas)):
    gamma = gammas[i]
    beta  = betas[i]
    
    # universal
    if gamma>0.8 or gamma<0.2:
        if gamma>0.8:
            centers[0] +=1
            total+=1
        else:
            centers[1] +=1
            total+=1  
    else:
        if gamma>0.5:
            if beta>0.35:
                centers[4]+=1
                total+=1
            elif beta<0.15:
                centers[2]+=1
                total+=1
        else:
            if beta>0.35:
                centers[3]+=1
                total+=1
            if beta<0.15:
                centers[5]+=1
                total+=1
  return fig, centers, total
```

```python
def get_these_graphs(percent, g_list = g_list):
  '''
  Gets graph indices in big_g_list that have that percent even degree nodes
  '''
  graphs_idx = []
  for i in range(len(g_list)):
    if even_percent(g_list[i])*100 == percent:
      graphs_idx.append(i) 
  return graphs_idx
```

```python
all_centers_ratios = [[] for i in range(6)]
percents = []
for p in range(0, 101, 10):
    percents.append(p)
    ind = get_these_graphs(p)
    fig , centers, total = get_centers(ind)
    print(p)
    print(total)
    centers_ratios = [i/total for i in centers]
    for j in range(6):
        all_centers_ratios[j].append(centers_ratios[j])
    
```
