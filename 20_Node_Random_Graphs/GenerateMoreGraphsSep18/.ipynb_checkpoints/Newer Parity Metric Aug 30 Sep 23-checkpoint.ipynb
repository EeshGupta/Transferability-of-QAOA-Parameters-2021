```python
file3 = open("20NodeRandomGraphs.txt")
mat_list = np.loadtxt(file3).reshape(55,20,20)
def mats_to_graphs(mat_list):
  """
  Given list of adjaceny matrix repr of graphs, converts to actual graphs
  """
  g_list = []
  for mat in mat_list:
    arr = np.array(mat)
    G = nx.convert_matrix.from_numpy_matrix(arr)
    g_list.append(G)
  return g_list
g_list = mats_to_graphs(mat_list)
graphs = g_list
```

```python
file1 = open("T1_local_55_20NodeGraphs(1).txt")
results  = np.loadtxt(file1).reshape(4,3025)
a = results[2]
b = results[3]
```

```python
def even_percent(graph):
  """
  Determines the relative percent of even degree nodes
  """
  nodes = graph.number_of_nodes()
  tot_ev = 0
  for i in range(nodes):
    if graph.degree[i]%2 == 0:
      tot_ev+=1
  return tot_ev/nodes

ev_deg_dict ={}
n = len(graphs)
for i in range(n):
  ev_deg_dict[i] = even_percent(graphs[i])
```

```python
def sort_for_plot(graphs = graphs, av =a,sub = b, ev_deg_dict = ev_deg_dict):
  """
  For plotting purposes, creates x,y,c lists
  """
  n = len(graphs)
  seeds = 20
  x = []
  y = []
  c = []
  d = []
  e = []
  for i in range(n):
    
    for j in range(n):
      x.append(ev_deg_dict[i])
      y.append(ev_deg_dict[j])
      print(n)
      r =  av[i*n + j]
      c.append(float(r))
      d_ = 1-0.34*(abs(ev_deg_dict[i] - ev_deg_dict[j]))
      d.append(d_)
      e.append(d_-r)
  return x, y, c,d,e
```

```python
%%capture
x, y ,c,d,e = sort_for_plot()
```

```python
def find_av_all(x, y, c, targx, targy):
    s = 0
    t=0
    for i  in range(len(x)):
        x_ = x[i]
        for j in range(len(y)):
            y_ = y[j]
            if (x_ == targx) and (y_ == targy) and (i==j):
                s+= c[i]
                t+=1
    #print(s)
    if t ==0:
        return 0
    return s/t
```
