```python
file3 = open("20NodeRandomGraphs.txt")
mat_list = np.loadtxt(file3).reshape(55,20,20)
def mats_to_graphs(mat_list):
  """
  Given list of adjaceny matrix repr of graphs, converts to actual graphs
  """
  g_list = []
  for mat in mat_list:
    arr = np.array(mat)
    G = nx.convert_matrix.from_numpy_matrix(arr)
    g_list.append(G)
  return g_list
g_list = mats_to_graphs(mat_list)
acc_list = g_list[:22]
```

```python
file4 = open("SmallRandomGraphs.txt")
data = []
for num in file4:
    data.append(float(num))
```

```python
graph_total_nodes =[ 2,  3,  3,  4,  4,  4,  5,  5,  5,  6,  6,  6,  6,  7,  7,  7,  7,
         8,  8,  8,  8,  8,  9,  9,  9,  9,  9, 10, 10, 10, 10, 10, 10]
```

```python
mat_list = []
c = 0
i = 0

while c< len(data):
    n = int(graph_total_nodes[i])
    n_sq = int(n**2)
    arr = np.reshape(data[c:c+n_sq], newshape = (n,n))
    mat_list.append(arr)
    
    c+= n_sq
    i+=1
```

```python
don_list = mats_to_graphs(mat_list)
len(don_list)
```

```python
file1 = open("T1_Small_To_Large.txt")
results  = np.loadtxt(file1).reshape(4,726)
a = results[2]
```

```python
def even_percent(graph):
  """
  Determines the relative percent of even degree nodes
  """
  nodes = graph.number_of_nodes()
  tot_ev = 0
  for i in range(nodes):
    if graph.degree[i]%2 == 0:
      tot_ev+=1
  return tot_ev/nodes

don_ev_deg_dict ={}
n = len(don_list)
for i in range(n):
  don_ev_deg_dict[i] = even_percent(don_list[i])

acc_ev_deg_dict ={}
n = len(acc_list)
for i in range(n):
  acc_ev_deg_dict[i] = even_percent(acc_list[i])
```

```python
x = []
y = []
z = []
for i in range(len(don_list)):
    for j in range(len(acc_list)):
        x.append(don_ev_deg_dict[i])
        y.append(acc_ev_deg_dict[j])
        z.append(a[i*len(acc_list) + j])
```

```python
cm = plt.cm.get_cmap('autumn')
sc = plt.scatter(x = x, y = y, vmax = 1,  c=z, cmap = cm)
plt.xlabel("% of even degree nodes in the donor")
plt.ylabel("% of even degree nodes in acceptor")
c_bar = plt.colorbar(sc)
```

```python
new_x=[]
new_y = []
for key in don_ev_deg_dict.keys():
    percentage =don_ev_deg_dict[key] 
    if percentage not in new_x:
        new_x.append(percentage)
        
for key in acc_ev_deg_dict.keys():
    percentage =acc_ev_deg_dict[key] 
    if percentage not in new_y:
        new_y.append(percentage)
```

```python
new_y
```

```python
def find_av_all(x, y, c, targx, targy):
    s = 0
    t=0
    for i  in range(len(x)):
        x_ = x[i]
        for j in range(len(y)):
            y_ = y[j]
            if (x_ == targx) and (y_ == targy) and (i==j):
                s+= c[i]
                t+=1
    if t ==0:
        return 0
    return s/t
            
```

```python
n_x = []
n_y = []
n_c = []
for x_ in new_x:
    for y_ in new_y:
        c_ = find_av_all(x, y, z, x_, y_)
        if c_ !=0:
            n_x.append(x_)
            n_y.append(y_)
            n_c.append(c_)
```

```python
cm = plt.cm.get_cmap('autumn')
sc = plt.scatter(x = n_x, y = n_y, vmax = 1,  c=n_c, cmap = cm)
c_bar = plt.colorbar(sc)
plt.xlabel("% of even degree nodes in the donor")
plt.ylabel("% of even degree nodes in acceptor")
```
